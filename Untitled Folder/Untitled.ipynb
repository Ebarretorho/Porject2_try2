```python
#import modules
import pandas as pd
import csv
```

```python
#append all individual michelin star(s) files into one bigd_data_set

path1michelin = "../data_sets_csv/one-star-michelin-restaurants.csv"
path2michelin = "../data_sets_csv/two-stars-michelin-restaurants.csv"
path3michelin = "../data_sets_csv/three-stars-michelin-restaurants.csv"

file1star = pd.read_csv(path1michelin)
file2star = pd.read_csv(path2michelin)
file3star = pd.read_csv(path3michelin)

file_1_2 = file1star.append(file2star)
file_1_2_3 = file_1_2.append(file3star)

file_1_2_3
```

```python
#Inspect the file in the region columns to see how are the restaurants distributed

file_1_2_3["region"].value_counts()
```

```python
#Since US located restaurants are located in different cities (in the region field), replace the region value from <city_name> to <"United States">

file_1_2_3["region"] = file_1_2_3["region"].replace({"Chicago":"United States", "Washington DC":"United States", "New York City":"United States", "California":"United States"})

file_1_2_3["region"].value_counts()
```

```python
#Create file data set with only restaurants within the United States region.
#Run a code to verify that we have effectively only kept US located restaurants

us_michelin_rest = file_1_2_3.loc[file_1_2_3["region"] == "United States" , : ]

us_michelin_rest["region"].value_counts()
```

```python
#Output the file into a csv for storage purposes.

us_michelin_rest.to_csv("../data_sets_csv/michelin_data_us.csv", index=False, header=True)
```
